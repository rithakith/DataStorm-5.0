```python
train_df = pd.read_csv("/kaggle/input/transactions/trx_table.csv")
```

```python
train_df.head()
```

```python
train_df.dtypes
```

```python
num_rows = len(train_df)
print(f"The number of rows in train_df is: {num_rows}")
```

Finding duplicate rows

```python
duplicates = train_df[train_df.duplicated()]

print("Duplicate rows:")
print(duplicates)
```

Removing duplicates

```python

train_df.drop_duplicates(inplace=True)

print("DataFrame after removing duplicates (in place):")
print(train_df)
```

```python
num_rows = len(train_df)
print(f"The number of rows in train_df is: {num_rows}")
```

Check for null values in all the columns

```python
null_counts = train_df.isnull().sum()

print("Null values in each column:")
print(null_counts)
```

Convert the 'transaction_time' column to datetime

```python
train_df['transaction_time'] = pd.to_datetime(train_df['transaction_time'])

print(train_df['transaction_time'].dtypes)
```

Get the min and max dates

```python
min_date = train_df['transaction_time'].min()
max_date = train_df['transaction_time'].max()

print(f"The minimum date in 'transaction_time' is: {min_date}")
print(f"The maximum date in 'transaction_time' is: {max_date}")
```

Load the item info CSV file

```python
item_info = pd.read_csv('/kaggle/input/item-information/item_info.csv')
```

Merge the DataFrames on the 'item code' column

```python
merged_df = pd.merge(train_df, item_info, on='item code', how='left')
```

```python
print(merged_df.head())
```

```python
num_rows = len(merged_df)
print(f"The number of rows in merged_df is: {num_rows}")
```

```python
merged_df.to_csv('merge.csv',index=False)
```
